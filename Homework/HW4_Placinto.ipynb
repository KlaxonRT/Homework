```
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np
 
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
 
model = Sequential()
model.add(Dense(2, input_dim=2))
#model.add(Activation('tanh')) # better than sigmoid this case
model.add(Activation('sigmoid'))
model.add(Dense(1))
model.add(Activation('sigmoid')) # tanh worked perfect, but there was an error returning neg prob.
 
sgd = SGD(lr=0.2) # stochastic gradient decent
model.compile(loss='mean_squared_error', optimizer=sgd)
#model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X, y, batch_size=1, epochs=1500)
print(model.predict(X))
print(model.predict(X).round())
```

```
# Plot corners first
xp = np.array([ [0], [0], [1], [1] ])
yp = np.array([ [0], [1], [0], [1] ])
plt.plot(xp, yp, 'ks')

for x in np.arange(0.0, 1.02, 0.02): # cannot use range since start and stop must be int
  for y in np.arange(0.0, 1.02, 0.02):
    temp = np.array([[x, y]])
    if model.predict(temp) > .5:
      plt.plot(x, y, 'b*')
    else:
      plt.plot(x, y, 'y.')
```
