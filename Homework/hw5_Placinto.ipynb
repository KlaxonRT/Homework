## HW5 - Guessing a Check Digit
x y z c, where c = (x+y+z) % 5  
Build the NN model to reach 6/7 = 85.7% test accuracy.

```python
# Austin Placinto
# I have neither given nor received any unauthorized aid in completing this work,
# nor have I presented someone else's work as my own.

import numpy as np
import keras
from keras import models
from keras import layers
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np

num_classes = 5

training_data = np.array([[0, 1, 2], 
                          [0, 7, 4],
                          [1, 4, 9], 
                          [1, 8, 6], 
                          [2, 3, 5],   
                          [2, 7, 0], 
                          [3, 2, 1], 
                          [3, 5, 8], 
                          [4, 3, 4], 
                          [4, 9, 2],                          
                          [5, 2, 8], 
                          [5, 6, 3], 
                          [6, 4, 0], 
                          [6, 7, 3], 
                          [7, 0, 8],                          
                          [7, 8, 6],
                          [8, 2, 8], 
                          [8, 6, 4], 
                          [9, 3, 1], 
                          [9, 8, 7],  
                         ], "float32")
train_labels = np.array([ [3], [1], [4], [0], [0], [4], [1], [1], [1], [0], [0], [4], [0], [1], [0], [1], [3], [3], [3], [4] ], "int")

# Unseen data for testing
test_data = np.array([ [1, 8, 6], 
                       [3, 2, 2],  
                       [4, 3, 4],  
                       [5, 6, 3],  
                       [6, 7, 3],  
                       [8, 8, 5], 
                       [9, 3, 1],                    
                     ],"float32")

test_labels = np.array([[0], [2], [1], [4], [1], [1], [3]], "int")

train_labels = keras.utils.to_categorical(train_labels, num_classes)
test_labels = keras.utils.to_categorical(test_labels, num_classes)

training_data = training_data/10 # to normalize from [0-9] to [0-1]
test_data = test_data/10
#print(training_data)

model = Sequential()
model.add(Dense(512, input_dim=3, activation='relu')) # relu was best. tanh or sigmod did not work...
model.add(Dense(64, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              #optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])

model.fit(training_data, train_labels, epochs=1000,
          validation_data = (test_data, test_labels),
          verbose=2)
          #validation_split = .1)
score = model.evaluate(training_data, train_labels, verbose = 0)
print(score[0])
print(score[1])

output = model.predict_classes(training_data)
print(output)
```

```python
accuracy = 0
model.summary()
test_labels = [0, 2, 1, 4, 1, 1, 3]
output = model.predict_classes(test_data)
print("Expected test labels:              [0 2 1 4 1 1 3]")
print("Output labels from trained model: ", output)
for x in range(len(test_labels)):
  if output[x] == test_labels[x]:
    accuracy+=1
print("Accuracy: ", (accuracy/len(test_labels))*100, "%")
```
